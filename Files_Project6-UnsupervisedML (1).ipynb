```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("diabetes.csv")
```

```python
data.head()
```

```python
import scipy.cluster.hierarchy as sch
plt.title("Dedrogram for Diabetes dataset")
dendrogram=sch.dendrogram(sch.linkage(data,method='ward'))
```
